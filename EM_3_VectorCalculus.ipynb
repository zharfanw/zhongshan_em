```python
import numpy as np
import scipy.integrate as spi
import matplotlib.pyplot as plt

# Parameterisasi lingkaran: x = cos(t), y = sin(t), t  [0, 2]
def x(t):
    return np.cos(t)

def y(t):
    return np.sin(t)

# Turunan parameterisasi
def dx_dt(t):
    return -np.sin(t)

def dy_dt(t):
    return np.cos(t)

# Medan vektor F = (P, Q) = (y, -x)
def P(x, y):
    return y

def Q(x, y):
    return -x

# Integran: P dx + Q dy
def integrand(t):
    return P(x(t), y(t)) * dx_dt(t) + Q(x(t), y(t)) * dy_dt(t)

# Hitung integral numerik
result, error = spi.quad(integrand, 0, 2*np.pi)

# --- Plot Lingkaran dan Medan Vektor ---
t_vals = np.linspace(0, 2*np.pi, 100)
x_vals = x(t_vals)
y_vals = y(t_vals)

# Grid untuk medan vektor
X, Y = np.meshgrid(np.linspace(-1.2, 1.2, 20), np.linspace(-1.2, 1.2, 20))
U = P(X, Y)  # Komponen x dari vektor
V = Q(X, Y)  # Komponen y dari vektor

plt.figure(figsize=(6,6))
plt.quiver(X, Y, U, V, color='gray', alpha=0.5)  # Plot medan vektor
plt.plot(x_vals, y_vals, 'b-', linewidth=2, label="Lintasan (Lingkaran)")  # Plot lingkaran
plt.scatter([0], [0], color='red', label="Pusat")  # Titik pusat lingkaran
plt.xlabel("x")
plt.ylabel("y")
plt.title(f"Closed Line Integral: {result:.4f}")
plt.legend()
plt.grid()
plt.axis("equal")  # Skala yang sama di x dan y
plt.show()
```

## 1.2. Surface Integrals

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.integrate import dblquad

# Parametrisasi bola
def r(u, v, R):
    return np.array([R * np.sin(u) * np.cos(v), R * np.sin(u) * np.sin(v), R * np.cos(u)])

# Turunan parsial terhadap u dan v
def ru(u, v, R):
    return np.array([R * np.cos(u) * np.cos(v), R * np.cos(u) * np.sin(v), -R * np.sin(u)])

def rv(u, v, R):
    return np.array([-R * np.sin(u) * np.sin(v), R * np.sin(u) * np.cos(v), 0])

# Menghitung norma dari produk cross (elemen luas)
def ds(u, v, R):
    cross_product = np.cross(ru(u, v, R), rv(u, v, R))
    return np.linalg.norm(cross_product)

# Integral untuk luas permukaan
def surface_area(R):
    integral, error = dblquad(lambda v, u: ds(u, v, R), 0, np.pi/2, lambda u: 0, lambda u: 2*np.pi)
    return integral

# Visualisasi bola
def plot_sphere(R):
    u = np.linspace(0, np.pi/2, 100)
    v = np.linspace(0, 2*np.pi, 100)
    u, v = np.meshgrid(u, v)
    x = R * np.sin(u) * np.cos(v)
    y = R * np.sin(u) * np.sin(v)
    z = R * np.cos(u)

    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(x, y, z, color='b', alpha=0.6)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    plt.title('Surface of Upper Hemisphere')
    plt.show()

R = 1  # Radius of the sphere
print("Calculated Surface Area:", surface_area(R))
plot_sphere(R)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from itertools import product, combinations

# Menghitung divergensi dari A
def divergence_A(x, y):
    return 2*x + 2*y

# Menghitung integral volumenya
def calculate_integral():
    x = np.linspace(0, 1, 100)
    y = np.linspace(0, 1, 100)
    dx = dy = 0.01  # Delta x dan delta y
    integral = 0
    for xi in x:
        for yi in y:
            integral += divergence_A(xi, yi) * dx * dy
    return integral

# Plot kubus
def plot_cube():
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    r = [0, 1]
    for s, e in combinations(np.array(list(product(r, r, r))), 2):
        if np.sum(np.abs(s-e)) == r[1]-r[0]:
            ax.plot3D(*zip(s, e), color="b")
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    plt.title('Cube from (0,0,0) to (1,1,1)')
    plt.show()

# Output
integral_value = calculate_integral()
print("Calculated Integral Value:", integral_value)
plot_cube()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

# Fungsi untuk menghitung komponen vektor A
def A(x, y, z):
    return np.array([x**2, y**2, -x*y])

# Mendefinisikan sisi kubus dan elemen luas dS
sides = {
    'positive x': (lambda y, z: A(1, y, z), np.array([1, 0, 0]), np.array([[1, 0], [0, 1]])),
    'negative x': (lambda y, z: A(0, y, z), np.array([-1, 0, 0]), np.array([[0, 0], [0, 1]])),
    'positive y': (lambda x, z: A(x, 1, z), np.array([0, 1, 0]), np.array([[0, 1], [0, 1]])),
    'negative y': (lambda x, z: A(x, 0, z), np.array([0, -1, 0]), np.array([[0, 1], [0, 0]])),
    'positive z': (lambda x, y: A(x, y, 1), np.array([0, 0, 1]), np.array([[0, 1], [0, 1]])),
    'negative z': (lambda x, y: A(x, y, 0), np.array([0, 0, -1]), np.array([[0, 1], [0, 1]]))
}

def integrate_surface():
    integral = 0
    for name, (A_func, normal, limits) in sides.items():
        A_integral = 0
        for x in np.linspace(limits[0][0], limits[0][1], 100):
            for y in np.linspace(limits[1][0], limits[1][1], 100):
                A_value = A_func(x, y)
                A_integral += np.dot(A_value, normal) * 0.01 * 0.01  # x * y
        integral += A_integral
        print(f"Flux through {name}: {A_integral}")
    return integral

# Plotting the cube
def plot_cube():
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    r = [-0.1, 1.1]
    for s, e in combinations(np.array(list(product(r, r, r))), 2):
        if np.sum(np.abs(s-e)) == r[1]-r[0]:
            ax.plot3D(*zip(s, e), color="b")
    plt.show()

integral = integrate_surface()
print("Total integral over the cube:", integral)
plot_cube()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib widget
# Konstanta
phi = 1  # Anda bisa menyesuaikan nilai ini sesuai dengan kebutuhan
r = 1    # Radius dari silinder

# Menghitung fluks melalui permukaan silinder
def calculate_flux():
    # Flux = integral dari (A . n) dS
    # Karena A = phi * r_hat dan n = r_hat pada permukaan silinder, A . n = phi * r
    # Luas permukaan silinder untuk dz dan dtheta (untuk r tetap) adalah r*dz*dtheta
    theta = np.linspace(0, 2 * np.pi, 100)
    z = np.linspace(-3, 3, 100)
    dtheta = theta[1] - theta[0]
    dz = z[1] - z[0]

    # Perhitungan integral sederhana
    flux = phi * r * (r * dtheta * dz) * len(theta) * len(z)
    return flux

# Plot silinder
def plot_cylinder():
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    z = np.linspace(-3, 3, 100)
    theta = np.linspace(0, 2 * np.pi, 100)
    theta, z = np.meshgrid(theta, z)
    x = r * np.cos(theta)
    y = r * np.sin(theta)
    
    ax.plot_surface(x, y, z, alpha=0.5, color='b')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    plt.title('Cylinder r=1, -3 <= z <= 3')
    plt.show()

flux = calculate_flux()
print("Calculated Flux:", flux)
plot_cylinder()
```

## 1.3. Volume integrals

```python
import numpy as np
from scipy.integrate import nquad
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definisi fungsi yang ingin diintegrasikan
def integrand(z, y, x):
    return x**2 + y**2 + z**2

# Batas integrasi untuk x, y, dan z
bounds = [0, 1]

# Menghitung integral
result, error = nquad(integrand, [bounds, bounds, bounds])
print(f"Hasil integral: {result}, dengan error: {error}")

# Membuat data untuk plot
x = np.linspace(0, 1, 100)
y = np.linspace(0, 1, 100)
x, y = np.meshgrid(x, y)
z = x**2 + y**2

# Membuat plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')

# Menampilkan plot
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('f(x, y, z) = x^2 + y^2')
ax.set_title('Visualisasi Fungsi f(x, y, z) = x^2 + y^2 + z^2')
plt.show()
```

```python
import numpy as np
from scipy.integrate import dblquad
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Fungsi untuk menghitung nilai z
def integrand(z, y):
    return np.sqrt(16 - y**2 - z**2)

# Batas integrasi untuk y dan z
y_bounds = (0, 2)
z_bounds = (-1, 1)

# Menghitung integral
volume, error = dblquad(integrand, y_bounds[0], y_bounds[1], 
                        lambda y: z_bounds[0], lambda y: z_bounds[1])

# Mengalikan dengan 2 untuk mencakup kedua sisi x (positif dan negatif)
total_volume = 2 * volume
print(f"Volume bagian bola: {total_volume} dengan error: {error}")

# Plot
phi = np.linspace(0, np.pi/2, 100)
theta = np.linspace(0, 2*np.pi, 100)
phi, theta = np.meshgrid(phi, theta)
x = np.sqrt(16) * np.sin(phi) * np.cos(theta)
y = np.sqrt(16) * np.sin(phi) * np.sin(theta)
z = np.sqrt(16) * np.cos(phi)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, color='b', alpha=0.3)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title('Visualisasi Bagian Bola')
plt.show()
```
