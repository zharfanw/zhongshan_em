```python
# Aktifkan mode interaktif untuk notebook
# %matplotlib notebook
# %matplotlib ipympl
%matplotlib widget
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Membuat figure dan sumbu 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Batas untuk sumbu
ax.set_xlim([0, 3])
ax.set_ylim([0, 3])
ax.set_zlim([0, 3])

# Label sumbu
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Vektor dx, dy, dz
dx = [1, 0, 0]
dy = [0, 1, 0]
dz = [0, 0, 1]

# Titik awal (origin)
origin = np.zeros(3)

# Menggambar vektor
ax.quiver(*origin, *dx, color='r', length=2, arrow_length_ratio=0.1)
ax.quiver(*origin, *dy, color='b', length=2, arrow_length_ratio=0.1)
ax.quiver(*origin, *dz, color='g', length=2, arrow_length_ratio=0.1)

ax.text(2, 0, -0.2, f"X", color='red')
ax.text(0, 2, -0.2, f"Y", color='blue')
ax.text(0, -0.2, 2, f"Z", color='green')

# Titik awal (origin)
origin_2 = np.array([1,1,1]);

# Menggambar vektor
ax.quiver(*origin_2, *dx, color='r', length=1, arrow_length_ratio=0.1)
ax.quiver(*origin_2, *dy, color='b', length=1, arrow_length_ratio=0.1)
ax.quiver(*origin_2, *dz, color='g', length=1, arrow_length_ratio=0.1)

# Menambahkan garis tebal dari (0,0,1) ke (0,0,2)
ax.plot([0, 0], [0, 0], [1, 2], 'm', linewidth=4)  # 'm' untuk warna magenta, garis tebal

# Menambahkan kubus untuk representasi
r = [0, 1]
X, Y = np.meshgrid(r, r)
Z = np.array([[1, 1], [1, 1]])
ax.plot_surface(X, Y, np.array([[1, 1], [1, 1]]), alpha=0.1, color='k', edgecolor='k')
ax.plot_surface(X, Y, np.array([[0, 0], [0, 0]]), alpha=0.1, color='k', edgecolor='k')
ax.plot_surface(X, np.array([[0, 0], [0, 0]]), Y, alpha=0.1, color='k', edgecolor='k')
ax.plot_surface(X, np.array([[1, 1], [1, 1]]), Y, alpha=0.1, color='k', edgecolor='k')
ax.plot_surface(np.array([[1, 1], [1, 1]]), X, Y, alpha=0.1, color='k', edgecolor='k')
ax.plot_surface(np.array([[0, 0], [0, 0]]), X, Y, alpha=0.1, color='k', edgecolor='k')

# Menampilkan plot
plt.show()
```
